```python
import pandas as pd
import re, os, pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# 1Ô∏è‚É£ Load data
df = pd.read_csv(r"C:\Users\Admin\oasis\spam_mail\data\spam.csv", encoding='latin-1')
df = df[['v1', 'v2']].rename(columns={'v1':'label','v2':'message'})
df['y'] = df['label'].map({'ham':0,'spam':1})

# 2Ô∏è‚É£ Split data
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['y'], test_size=0.2, random_state=42)

# 3Ô∏è‚É£ Vectorize
vectorizer = TfidfVectorizer(stop_words='english', lowercase=True, max_features=3000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 4Ô∏è‚É£ Train
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# 5Ô∏è‚É£ Evaluate
print("‚úÖ Accuracy:", accuracy_score(y_test, model.predict(X_test_vec)))
print(classification_report(y_test, model.predict(X_test_vec)))

# 6Ô∏è‚É£ Save model + vectorizer
os.makedirs(r"C:\Users\Admin\oasis\spam_mail\model", exist_ok=True)
pickle.dump(model, open(r"C:\Users\Admin\oasis\spam_mail\model\spam_model.pkl", "wb"))
pickle.dump(vectorizer, open(r"C:\Users\Admin\oasis\spam_mail\model\vectorizer.pkl", "wb"))

print("üéØ Model and vectorizer saved successfully!")
```
